```
import itertools

alphabet = input()
n = input()
combs = []
for i in range(1, n + 1):
    combs.append(list(map(''.join, (itertools.product(alphabet, repeat=i)))))
combs = list(itertools.chain(*combs))
srt_combs = sorted(combs, key=lambda word: [alphabet.index(c) for c in word])
dist = srt_combs.index(alphabet[1]) - srt_combs.index(alphabet[0])
for i in range(len(alphabet)):
    j = srt_combs.index(alphabet[i])
    print(srt_combs[j:j+dist])
```

```
coins = input()
coins.sort(reverse = True)
price = input()
coin_quantity = {}
for i in coins:
    coin_quantity[i] = price / int(i)
    price = price % i
    if price == 0: break
print(coin_quantity)
```
