```
lst = []
n = int(input())
a = []
for i in range (n):
  l, r = map(int, input().split())
  lst.append([l,r])
lst.sort(key = lambda s: s[1])
while lst:
  a.append(lst[0][1])
  lst = [lst[i] for i in range(len(lst)) if lst[i][0] > lst[0][1]]
print(len(a), str(a))
```

```
n, v = map(int, input().split())
lst = []
for i in range(n):
  c, w = map(int, input().split())
  lst.append([c, w, c/w])
lst.sort(key = lambda s: s[2], reverse = True)
m = 0
while v != 0:
  if lst[0][1] <= v:
      m += lst[0][2] * lst[0][1]
      v -= lst[0][1]
      del lst[0]
  else: m += lst[0][2] * v
print (str('%.3f' % m))
      
```

```
import heapq
from collections import Counter, namedtuple


class Node (namedtuple("Node", ["left", "right"])): # класс Нод, внутренний узел
            def walk(self, code, acc):
                self.left.walk(code, acc + "0")
                self.right.walk(code, acc + "1")
            
            
class Leaf (namedtuple("Leaf", ["char"])): # лист, хранящий символ
            def walk(self, code, acc):
                code[self.char] = acc or "0"


def huffman_encode(s):
  h = []
# h = [(freq, Leaf(ch)) for ch, freq in Counter(s).items()] # каждый лист с новым символом и его частота заносим в очередь
  for ch, freq in Counter(s).items():
    h.append((freq, len(h), Leaf(ch)))
    
  heapq.heapify(h) # создание очереди с приоритетами по частотам
  
  count = len(h)
  while len(h) > 1:
    freq1, _count1, left = heapq.heappop(h) # 1й символ с наименшим приоритетом
    freq2, _count2, right = heapq.heappop(h) # 2й символ с наименьшим приоритетом
    heapq.heappush(h, (freq1 + freq2, count, Node(left, right))) # добавляем узел с 2мя детьми в очередь
    count += 1

  code = {}
  if h: 
    [(_freq, count, root)] = h # корень дерева
    root.walk(code, "")
    
  return code


def huffman_decode(s, code):
  s0 =[ch for ch in s] # список символов
  s1 = "" # декодированная строка
  while s0:
    if s0[0] in code.values(): # если в словаре содержится ключ с таким значением
      for l, cod in code.items():
        if cod == s0[0]: s1 += l #приписываем к д-строке ключ-букву
      del s0[0] # удаляем расшифрованный кусок кода
    else:
      s0[0] = (s0[0])+(s0[1]) 
      del s0[1] 
  return s1  
  
  
def test(n_iter = 100):
  import random
  import string
  
  for i in range(n_iter):
    length = random.randint(0,32)
    s = "".join(random.choice(string.ascii_letters) for _ in range(length))
    code = huffman_encode(s)
    encoded = "".join(code[ch] for ch in s)
    assert s == huffman_decode(encoded, code)
    print(s)
    print(s == huffman_decode(encoded, code), '\n')

  

def main():
  s = input()
  code = huffman_encode(s)
  encoded = "".join(code[ch] for ch in s)
  print(len(code), len(encoded))
  for ch in sorted(code):
    print("{}: {}".format(ch, code[ch]))
  print(encoded)
  

if __name__ == "__main__":
  test(3)
```
